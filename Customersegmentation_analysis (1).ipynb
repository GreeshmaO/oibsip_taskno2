```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X = df[['Income', 'MntTotal']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**ELBOW METHOD:**
The elbow method is a technique used to determine the optimal number of clusters in a dataset when performing clustering analysis, such as k-means clustering. It involves plotting the within-cluster sum of squares (WCSS) against the number of clusters, and identifying the "elbow point" where the rate of decrease in WCSS slows down significantly. This point indicates the optimal number of clusters.


1.Choose a Range of K:
Start by selecting a range of values for the number of clusters (k) to be tested. Typically, you'll choose a range from 1 to some maximum value.

2.Apply K-means:
Apply the k-means clustering algorithm to the dataset for each value of k in the chosen range.

3.Compute WCSS:
For each value of k, calculate the within-cluster sum of squares (WCSS), which is the sum of squared distances between each data point and its nearest centroid.

4.Plot the Elbow Curve:
Plot the number of clusters (k) against the corresponding WCSS values.
The plot will typically show a decreasing trend in WCSS as k increases.

5.Identify the Elbow Point:
Identify the "elbow point" in the plot, where the rate of decrease in WCSS slows down significantly.

The elbow point is the optimal number of clusters, as adding more clusters beyond this point may not significantly reduce the WCSS.

6.Choose the Optimal K:
Choose the number of clusters corresponding to the elbow point as the optimal value of k for clustering the dataset.

```python

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Based on the Elbow Method, select the optimal number of clusters
n_clusters = 5
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
kmeans.fit(X_scaled)
labels = kmeans.labels_
```

```python

# Add cluster labels to the original DataFrame
df['Cluster'] = labels
```

**Visualize the clusters:**

```python

plt.figure(figsize=(10, 6))
for i in range(n_clusters):
    plt.scatter(X.loc[labels == i, 'Income'], X.loc[labels == i, 'MntTotal'], label=f'Cluster {i}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.title('Customer Segmentation')
plt.xlabel('Income')
plt.ylabel('MntTotal')
plt.legend()
plt.show()
```

```python
# Dimensionality reduction for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Plot clusters in 2D
plt.figure(figsize=(10, 6))
for i in range(n_clusters):
    plt.scatter(X_pca[labels == i, 0], X_pca[labels == i, 1], label=f'Cluster {i}')
plt.scatter(pca.transform(kmeans.cluster_centers_)[:, 0], pca.transform(kmeans.cluster_centers_)[:, 1], s=300, c='red', label='Centroids')
plt.title('Customer Segmentation (PCA)')
plt.xlabel('Income')
plt.ylabel('MntTotal')
plt.legend()
plt.show()
```
